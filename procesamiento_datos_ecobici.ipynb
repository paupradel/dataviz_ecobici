### Procesamiento de viajes de Ecobici

__Descripción:__

Se obtiene el AGEB de origen y destino de los viajes de los viajes de ecobici y se obtiene el agregado primedio de los viajes entre AGEBS por hora.

__Input__

- Estaciones de ecobici: Estaciones de ecobici con clave AGEB obtenida de __Uber © 2019 Copyright Uber Technologies, Inc.Data Attributions__

- Viajes Ecobici 2019
Responsable: Ecobici



```python
import pandas as pd
```

## Estaciones

```python
df_estaciones = pd.read_csv('./SIG/data/procesado/CSV/estaciones-de-ecobici_con_AGEB.csv',sep=',')
df_estaciones.head()
```

## Viajes ecobici

```python
df_viajes = pd.read_csv('./data/2019/ecobici_2019.csv',sep=',')
df_viajes.head()
```

### merge viajes e info de estaciones

Se obtiene ageb por estación de origen y destino.

#### Clave AGEB arribo

```python
df_viajes_ageb = df_viajes.merge(df_estaciones[['ID','CVE_AGEB']], left_on='Ciclo_Estacion_Arribo', right_on='ID')
```

```python
df_viajes_ageb.rename(columns={'CVE_AGEB': 'CVE_AGEB_arribo'}, inplace=True)
```

```python
df_viajes_ageb.head()
```

#### Clave AGEB destino

```python
df_viajes_ageb = df_viajes_ageb.merge(df_estaciones[['ID','CVE_AGEB']], left_on='Ciclo_Estacion_Retiro', right_on='ID')
```

```python
df_viajes_ageb.rename(columns={'CVE_AGEB': 'CVE_AGEB_retiro'}, inplace=True)
```

```python
df_viajes_ageb.head()
```

### Eliminamos columnas no usadas

```python
df_viajes_ageb.columns
```

```python
df_viajes_ageb.drop(['Unnamed: 9','ID_x','ID_y'], axis=1, inplace=True)
```

```python
df_viajes_ageb.head()
```

## Obtenemos la columna de hora de retiro:

```python
df_viajes_ageb['Hora_Retiro_round'] = df_viajes_ageb['Hora_Retiro'].str.split(':').str[0].astype(int)
```

```python
df_viajes_ageb.head()
```

## Obtenemos el tiempo de cada viaje:

Eliminamos filas con errores de formato:

```python
df_viajes_ageb[df_viajes_ageb['Hora_Retiro']=='18::']
```

```python
df_viajes_ageb.drop([3127563],inplace=True)
```

Convetimos a datetime:

```python
df_viajes_ageb['Hora_Retiro'] = pd.to_datetime(df_viajes_ageb['Hora_Retiro'], format='%H:%M:%S')

df_viajes_ageb['Hora_Arribo'] = pd.to_datetime(df_viajes_ageb['Hora_Arribo'], format='%H:%M:%S')
```

Verificamos que las horas de arribo y reutro sean datetime

```python
df_viajes_ageb.dtypes
```

restamos las columnas para obtener el tiempo

```python
df_viajes_ageb['duracion_viaje'] = df_viajes_ageb['Hora_Arribo'] - df_viajes_ageb['Hora_Retiro']
```

```python
df_viajes_ageb['duracion_viaje_minutos'] = df_viajes_ageb['duracion_viaje'].astype('str').str.split(':').str[1].astype(int)
```

```python
df_viajes_ageb.head()
```

```python
df_viajes_ageb.drop(['Bici','Ciclo_Estacion_Arribo','Ciclo_Estacion_Retiro','Fecha_Arribo',
                    'Fecha_Retiro'], axis=1, inplace=True)
```

```python
df_viajes_ageb.to_csv('./data/procesado/datos_ecobici_ageb.csv',index=False)
```

## Groupby de tiempo:

Obtenemos el tiempo y edad para de viajes entre dos agebs por hora del viaje:

```python
df_viajes_ageb_agg = df_viajes_ageb.groupby(['CVE_AGEB_arribo','CVE_AGEB_retiro','Hora_Retiro_round']).agg({'Edad_Usuario':['count', 'mean'], 
                         'duracion_viaje_minutos':['mean','min','max']}).reset_index(drop=False)

df_viajes_ageb_agg.columns = ['_'.join(col) for col in df_viajes_ageb_agg.columns]
```

Convirtamos colunnas a __int__:

```python
colums = ['Edad_Usuario_count', 'Edad_Usuario_mean',
       'duracion_viaje_minutos_mean', 'duracion_viaje_minutos_min',
       'duracion_viaje_minutos_max']

df_viajes_ageb_agg[colums] = df_viajes_ageb_agg[colums].astype(int)
```

```python
df_viajes_ageb_agg.head()
```

```python
df_viajes_ageb_agg.rename(columns={'Edad_Usuario_count':"Número_de_Viajes"},inplace=True)
```

```python
df_viajes_ageb_agg.head()
```

```python
df_viajes_ageb_agg.to_csv('./data/procesado/datos_ecobici_ageb_groupby_agg.csv',index=False)
```
