```python
dfbn = pd.read_csv('/home/pradel/Desktop/ecobici_viz/data/production_data/viajes_ecobici.csv')
```

```python
viajes_ecobici = dfbn[['CVE_AGEB_arribo_', 'CVE_AGEB_retiro_']].drop_duplicates(subset=['CVE_AGEB_arribo_', 'CVE_AGEB_retiro_']).reset_index(drop=True)
viajes_ecobici.head()
```

```python
row_trip = 2246

ageb_arribo = viajes_ecobici.loc[row_trip,"CVE_AGEB_arribo_"]
ageb_retiro = viajes_ecobici.loc[row_trip,"CVE_AGEB_retiro_"]
print(ageb_retiro, ageb_arribo)
```

```python
dfbn_filter = dfbn[(dfbn['CVE_AGEB_retiro_'] == ageb_retiro) & 
            (dfbn['CVE_AGEB_arribo_'] == ageb_arribo)]
```

```python
dfbn_filter.head()
```

```python
edad_promedio_usuario= int(round(dfbn_filter['Edad_Usuario_mean'].mean()))
```
