```python
geodf.to_file('test_uber_json.json', driver= 'GeoJSON')
with open ('test_uber_json.json') as geofile:
    jdata=json.load(geofile)
```

```python
def check_geojson(jdata):
    if 'id' not in jdata['features'][0].keys():
        if 'properties' in jdata['features'][0].keys():
            if 'id' in jdata['features'][0]['properties'] and jdata['features'][0]['properties']['id'] is not None:
                for k, feat in enumerate(jdata['features']):
                    jdata['features'][k]['id']= feat['properties']['id']
            else:
                for k in range(len(jdata['features'])):
                    jdata['features'][k]['id']= k
    return jdata
```

```python
z = geodf['790'].tolist()
```

```python
locations = geodf["CVE_AGEB"].tolist()
```

```python
jdata = check_geojson(jdata)
```

```python
jdata
```

```python
trace = go.Choroplethmapbox(z = z,
                            locations=locations,
                            colorscale= 'Viridis',
                            geojson=jdata,
                            hoverinfo= 'all',
                            marker_line_width=0.1, marker_opacity=0.5)
```

```python
trace
```

```python
layout= go.Layout(title_text= 'Choroplethmapbox',
                   title_x=0.5, width = 700, height=700,
                   mapbox = dict(center= dict(lat=19.410737,  lon=-99.170879),
                                 style= 'carto-positron',
                                 zoom=11.5,
                               ))
```

```python
fig= go.Figure(data=[trace], layout=layout)
fig.update_layout(mapbox_style= 'carto-positron')
fig.show()
```

```python
df_uber = pd.read_csv('/home/pradel/Desktop/ecobici_viz/data/production_data/viajes_uber.csv')
```

```python
dfbn['Hora_Retiro_bin']=dfbn['Hora_Retiro_bin_']
```

```python
df_uber.head()
```

```python
dfbn.head()
```

```python
def asignar_bote(df, columna= 'Hora_Retiro_bin'):
    if df[columna]=='7-10':
        return 0
    if df[columna]=='10-13':
        return 1
    if df[columna]=='13-16':
        return 2
    if df[columna]=='16-19':
        return 3
    if df[columna]=='19-22':
        return 4
    if df[columna]=='22-1':
        return 5
```

```python
df_uber['periodo']= df_uber.apply(asignar_bote, axis=1)
```

```python
dfbn['periodo']= dfbn.apply(asignar_bote, axis=1)
```

```python
df_uber_filter = df_uber[(df_uber['sourceid'] == 1375) & 
            (df_uber['dstid'] == 292)]
```

```python
dfbn_filter= dfbn[(dfbn['CVE_AGEB_retiro_'] == 1375) & 
            (dfbn['CVE_AGEB_arribo_'] == 292)]
```

```python
df_uber_filter= df_uber_filter.sort_values(by='periodo', ascending=True)
```

```python
dfbn_filter= dfbn_filter.sort_values(by='periodo', ascending=True)
```

```python
x_uber=df_uber_filter['Hora_Retiro_bin']
```

```python
y_uber=df_uber_filter['mean_travel_time']
```

```python
x_ecobici=dfbn_filter['Hora_Retiro_bin']
```

```python
y_ecobici=dfbn_filter['duracion_viaje_minutos_mean']
```

```python
data_bar= [go.Bar(name= 'ecobici',x=x_ecobici,y=y_ecobici),
          go.Bar(name= 'uber', x=x_uber, y=y_uber)]
```

```python
layout_bar= go.Layout(xaxis= {'type': 'category'})
```

```python
figure_bar= go.Figure(data=data_bar, layout= layout_bar)
```
